```python
# Import Dependencies
import pandas as pd

import psycopg2
from config import password

import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
```

```python
from sqlalchemy import create_engine
engine = create_engine(f'postgresql://postgres:{password}@localhost:5432/sqlhomework')
connection = engine.connect()
```

```python
employee_df = pd.read_sql('select * from employees', connection)
employee_df
```

```python
salaries_df = pd.read_sql('select * from salaries', connection)
salaries_df
```

```python
salaries_df.min()
```

```python
salaries_df.max()
```

```python
emp_salary = pd.merge(employee_df, salaries_df, on = 'emp_no')
emp_salary2 = emp_salary.rename(columns={"emp_title":"title_id"})
emp_salary2
```

```python
x = emp_salary2['salary']
plt.hist(x, bins=15)
plt.show()
```

```python
titles_df = pd.read_sql('select * from titles', connection)
titles_df
```

```python
emp_salary_title = pd.merge(emp_salary2, titles_df, on = "title_id")
emp_salary_title
```

```python
title_avg = emp_salary_title.groupby(['title']).mean()['salary'] 
title_avg.round(2)
```

```python
x_axis = title_avg.index.values
x_axis
```

```python
y_axis = title_avg.values
y_axis
```

```python
#bar chart using pyplot
plt.bar(x_axis, y_axis)
plt.xticks(rotation = 45)
plt.title("Average Salary per Job Title")
plt.xlabel("Job Title")
plt.ylabel("Salary ($)")
plt.show()
```
